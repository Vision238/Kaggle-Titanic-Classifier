```python
import numpy as np
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
mean = train["Age"].mean()
std = train["Age"].std()
mean
train["Age"].fillna(mean,inplace=True)
train["Age"] = train["Age"].astype(int)
y = train["Survived"]
```

```python
type(train)
```

```python
train_data = train[["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
train_data.head()
```

```python
clf = RandomForestClassifier(random_state=0)
X = pd.get_dummies(train_data)
clf.fit(X,y)
```

```python
mean = train["Age"].mean()
std = train["Age"].std()
mean
test["Age"].fillna(mean,inplace=True)
test["Age"] = test["Age"].astype(int)
```

```python
mean = train["Fare"].mean()
std = train["Fare"].std()
mean
test["Fare"].fillna(mean,inplace=True)
test["Fare"] = test["Fare"].astype(int)
```

```python
test_data = test[["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
X_test = pd.get_dummies(test_data)
```

```python
test_data.describe()
```

```python
predictions = clf.predict(X_test)
```

```python
result = pd.DataFrame({"PassengerId":test["PassengerId"],"Survived":predictions})
```

```python
result.head()
```

```python
result.to_csv('submission.csv', index=False)
```
